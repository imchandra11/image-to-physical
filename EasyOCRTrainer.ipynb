```python
"""This script tests the EasyOCRMetricsManager by providing sample ground truth and predicted text data.
   It computes and prints the Word Error Rate (WER) and Character Error Rate (CER) to evaluate OCR performance."""


from utils.easyocrmetrics import EasyOCRMetricsManager
ground_truths = ["hello world", "deep learning is powerful", "optical character","recognition","easyocr is working well","text extraction from images"]
predictions = ["hello word","deep learning powerful","optical character"," recogniton","easyocr is workng well","text extraction images"]
metrics = EasyOCRMetricsManager()
metrics.update(ground_truths, predictions)
result = metrics.summarize()
print("===== EasyOCR Evaluation Results =====")
print(f"Word Error Rate (WER): {result['wer']:.4f}")
print(f"Character Error Rate (CER): {result['cer']:.4f}")
```
